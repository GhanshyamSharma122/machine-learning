## Understanding k-Means Clustering

k-Means clustering is an unsupervised machine learning algorithm used to partition `n` observations into `k` clusters, where each observation belongs to the cluster with the nearest mean (centroid), serving as a prototype of the cluster.

**How it works:**
1.  **Initialization**: Randomly select `k` data points from the dataset as initial cluster centroids.
2.  **Assignment Step**: Assign each data point to its closest centroid, forming `k` clusters.
3.  **Update Step**: Recalculate the centroids of the newly formed clusters (usually by taking the mean of all data points within each cluster).
4.  **Repeat**: Steps 2 and 3 are repeated until the centroids no longer change significantly, or the maximum number of iterations is reached.
